### 1.What is the total population of India according to the dataset?

```python
import pandas as pd
df = pd.read_csv('6. India Census 2011.csv')
df.head()
```

```python
a= df['Population'].sum()
print(f"The total population of India as per 2011 Census is {a}.")
```

### 3. What is the Total population across all states/UTs?

```python
df1 = df.groupby(by = 'State_name').agg({'Population': 'sum'}).reset_index()
df1.sort_values(by= 'Population', ascending =  False, inplace=True)
df1['Population %'] = round((df1['Population']/a)*100,3)
df1
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
abc = df1.iloc[0:10,2].tolist()
lab = df1.iloc[0:10,0].tolist()
plt.pie(x= abc,labels = lab, autopct='%1.1f%%')
plt.title('Pie Chart showing Popualtion % of Top 10 Highest Population State/UT')
plt.show()
```

### 15. Perform a correlation analysis between literacy rate and education levels (Secondary, Higher, and Graduate Education).

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df9 = df[['State_name', 'District_name', 'Population', 'Literate',
       'Secondary_Education', 'Higher_Education', 'Graduate_Education']]
df9
```

```python
sns.heatmap(df9.corr(), annot= True)
plt.show()
```

#### Key Insights from the Heatmap:
##### Correlation Between Literacy Rate and Education Levels:

- Literate vs Secondary_Education â†’ 0.93
â†’ Strong positive correlation. More literacy generally leads to more secondary education.
- Literate vs Higher_Education â†’ 0.92
â†’ Slightly weaker than secondary education but still a strong correlation.
- Literate vs Graduate_Education â†’ 0.87
â†’ Moderate correlation, suggesting that while literacy contributes to graduate education, other factors may also play a role.

##### Interpretation:

Literacy strongly influences access to Secondary and Higher Education.
Graduate Education shows a slightly weaker correlation, implying that other socio-economic factors might come into play.
Secondary Education (0.93) has the highest correlation with literacy, meaning it benefits the most from literacy improvements.

##### Conclusion:
Improving literacy rates will most directly increase Secondary and Higher Education enrollment.
Graduate education is still correlated with literacy but is more dependent on other factors like economic conditions, affordability, and personal choices.

### 17. How does the workforce participation rate vary between male and female populations across states?

```python
df10 = df[['District_code', 'State_name', 'District_name', 'Population', 'Male',
       'Female', 'Workers', 'Male_Workers', 'Female_Workers']]
df10 = df10.groupby(by ='State_name').agg({'Population':'sum','Male':'sum','Female':'sum','Workers':'sum','Male_Workers':'sum','Female_Workers':'sum'}).reset_index()
df10
```

```python
df10['WFPR'] = round((df10['Workers']/df10['Population'])*100,2)
df10['male_WFPR'] = round((df10['Male_Workers']/df10['Male'])*100,2)
df10['female_WFPR'] = round((df10['Female_Workers']/df10['Female'])*100,2)
df10.sort_values(by= 'male_WFPR', ascending =False,inplace = True)
df10
```

```python
highest_wfpr_state = df10.loc[df10['WFPR'].idxmax()]

highest_male_wfpr_state = df10.loc[df10['male_WFPR'].idxmax()]

highest_female_wfpr_state = df10.loc[df10['female_WFPR'].idxmax()]

print("State with highest WFPR:", highest_wfpr_state['State_name'])
print("State with highest male WFPR:", highest_male_wfpr_state['State_name'])
print("State with highest female WFPR:", highest_female_wfpr_state['State_name'])
```
